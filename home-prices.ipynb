## Importing Packages

```python
import  pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Reading Data

```python
train = pd.read_csv("home/train.csv", na_values=['NaN'])

print('Total Features:', len(train.columns.values))
print('Total Records:', train.shape[0])
```

## Printing Sample Data

```python
train.tail()
```

## Analyzing Column Data

```python
train.describe()
```

```python
train.describe(include=['O'])
```

```python
# Finding all the columsn who have at least 10% of values as null
null_col_list = train.columns[train.isnull().sum() > 146].tolist()

print(null_col_list)
```

- As seen above there are large number of features, humanly it will be time consuming to analyze each feature with respect to sale price. Hence, we will use seaborn package to plot a heatmap of the correlation between the features.
- In order to correlate we need to convert all non-numeric data into numeric data. Let us first find out all the numeric columns.
- Also we will be eliminating those columns who have at least 10% of values as null/


## Eliminating Null Valued Columns

```python
train = train.drop(null_col_list, 1)

print('Number of Cols:', len(train.columns))
```

```python
# Non Numeric Column List
non_numeric_cols = train.select_dtypes(exclude=['number'])

non_numeric_cols.columns.values
```

## Converting Non Numeric Columns
- On the basis of data description given in the channel we will map the non-numeric data to numeric values

```python
# Column Mapping
for col in non_numeric_cols:
    unique_vals = train[col].unique()
    
    col_dict = {}
    # Mapping the unique values with numeric value
    for i in range(0, len(unique_vals)):
        col_dict[unique_vals[i]] = i
        
    train[col] = train[col].map(col_dict).astype(int)
```

## Plotting Heatmap
- Now we have converted all the columns into numeric data, hence let us plot the heatmap

```python
corr = train.dropna().corr()
sns.set(font_scale=1)
sns.heatmap(corr, vmax=1, square=True)
plt.show()
```

- As we are interested in the SalePrice let us see the correlation of Sale Price with other features

```python
pd.options.display.max_rows = 4000
corr_list = corr['SalePrice'].sort_values(ascending=False)[1:]
print(corr_list)
```

- In the above th list we can see the various correlation values, higher the value more closely is the feature related with SalePrice.
- Let us pick up all the columns which have correlation value greater than 0.5 and experiment

```python
corr_list.index
```

```python
feature_cols = []

for i in range(0, len(corr_list)):
    if corr_list[i] >= 0.5:
        feature_cols.append(corr_list.index[i])

feature_cols = feature_cols      
print(feature_cols)
```

```python
# Extracting feature columns from the dataset
train = train[feature_cols + ['SalePrice']]

train.tail()
```

```python
# Checking for the Null values in the selected columns
train.isnull().sum()
```

- There are no null values in our newly created dataset, hence let us proceed with training of the dataset

```python
train.describe()
```

## Preparing Features and Labels

```python
features = train[feature_cols].values
labels = train['SalePrice'].values
```

## Training
- We will be using regression to train our model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

# Splitting Training Data

```python
# Splitting training dataset into testing and training set
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.1, shuffle=True)
```

```python
# Defining Model
model = LinearRegression()

# Fitting data
model.fit(X_train, y_train)
```

## Testing Data

```python
y_predict = model.predict(X_test)
print('Score:', r2_score(y_true=y_test, y_pred=y_predict))
```

## Outlier Removal
- As seen above we have low r2 score and hence we reccomend to remove the outlier data to improve the score

```python
n = 5

new_features = features
new_labels = labels

for a in range(0, n):
    # Calculating residual error
    predict = model.predict(new_features)
    res = abs(predict - new_labels)
    
    k = (int)(0.1 * len(res))

    temp_features = []
    temp_labels = []

    thresh = np.sort(res)[::-1][k]

    for i in range(0, len(res)):
        if res[i] < thresh:
            temp_features.append(new_features[i])
            temp_labels.append(new_labels[i])
    
    new_features = temp_features
    new_labels = temp_labels
    
    X_train, X_test, y_train, y_test = train_test_split(new_features, new_labels, test_size=0.10, shuffle=True)
    
    model.fit(X_train, y_train)
    
    y_predict = model.predict(X_test)
    print('Score:', r2_score(y_true=y_test, y_pred=y_predict))
```

## Working on Test Data

```python
test = pd.read_csv('home/test.csv', na_values=['NaN'])
```

```python
feature_df = test[feature_cols]
data_ids = test['Id'].values
```

## Analyzing for Null Values

```python
feature_df.isnull().sum()
```

- As we can see that there are certain values which are null, let us subsitute with average value

```python
feature_df['GarageCars'] = feature_df['GarageCars'].fillna((int)(train['GarageCars'].mean())).astype(int)
feature_df['GarageArea'] = feature_df['GarageArea'].fillna((int)(train['GarageArea'].mean())).astype(int)
feature_df['TotalBsmtSF'] = feature_df['TotalBsmtSF'].fillna((int)(train['TotalBsmtSF'].mean())).astype(int)
```

```python
features = feature_df.values

labels = model.predict(features)
```

```python
data = np.transpose(np.vstack((data_ids, labels)))
data = data.astype(np.int)



import csv

with open('result-home.csv', 'w', newline='') as f:
    writer = csv.writer(f)

    writer.writerows([['Id', 'SalePrice']])
    writer.writerows(data)
```
